```python
#Qi en Sam 6
years = list(set(df['iyear']))

df_budget = pd.read_csv('military-expenditure-as-share-of-gdp.csv',  encoding = "ISO-8859-1")
us_budget = df_budget[df_budget['Code'] == 'USA']
us_budget = us_budget[us_budget['Year'] > 1970]


world_budget = [np.average(df_budget[df_budget['Year'] == year]['Expenditure']) for year in years]
df_budget.head(n=10)


# Create a trace
budget_trace = go.Scatter(
    x = us_budget['Year'],
    y = world_budget,
    yaxis='y2',
    name="Global spending"
)


deaths_per_year = [np.nansum(df[df['iyear'] == year]['nkill']) for year in years]


data = go.Bar(
    name="Deaths",
    x=years,
    y=deaths_per_year
)

layout = go.Layout(
    title='Number of deaths by terrorist attacks per year',
    yaxis=dict(
        title='Number of deaths'
    ),
    yaxis2=dict(
        title='World budget as percent of gpb',
        overlaying='y',
        side='right'
    )
)



fig = go.Figure(data=[data, budget_trace], layout=layout)
iplot(fig)
```

```python
#Thomas World map

#kills per country
kpc = df.groupby("country_txt").sum()

#get country names
countries = kpc.index.get_level_values(0)
#country lattitude and longitude
cll = df.groupby("country_txt").median()


scl = [0,"rgb(150,0,90)"],[0.125,"rgb(0, 0, 200)"],[0.25,"rgb(0, 25, 255)"],\
[0.375,"rgb(0, 152, 255)"],[0.5,"rgb(44, 255, 150)"],[0.625,"rgb(151, 255, 0)"],\
[0.75,"rgb(255, 234, 0)"],[0.875,"rgb(255, 111, 0)"],[1,"rgb(255, 0, 0)"]

data = [go.Scattergeo(
    lat = cll["latitude"],
    lon = cll["longitude"],
    text = kpc['nkill'].astype(str) + " deaths in " + countries.astype(str),
    marker = dict(
        color = kpc['nkill'],
        colorscale = scl,
        reversescale = False,
        opacity = 0.7,
        size = np.log(kpc['nkill'], where=kpc['nkill']>0)*4,        
        colorbar = dict(
            thickness = 10,
            titleside = "right",
            outlinecolor = "rgba(68, 68, 68, 0)",
            ticks = "outside",
            ticksuffix = " deaths",
        )                
    )
)]

layout = dict(
    geo = dict(
        scope = 'world',
        showland = True,
        landcolor = "rgb(212, 212, 212)",
        subunitcolor = "rgb(255, 255, 255)",
        countrycolor = "rgb(255, 255, 255)",
        showlakes = True,
        lakecolor = "rgb(255, 255, 255)",
        showsubunits = True,
        showcountries = True,
        resolution = 50,
        projection = dict(
            type = 'conic conformal',
            rotation = dict(
                lon = -100
            )
        ),
        lonaxis = dict(
            showgrid = True,
            gridwidth = 0.5,
            range= [ -140.0, -55.0 ],
            dtick = 5
        ),
        lataxis = dict (
            showgrid = True,
            gridwidth = 0.5,
            range= [ 20.0, 60.0 ],
            dtick = 5
        )
    ),
    title = 'Attacks resulting in more than 5 deaths',
)

fig = go.Figure(data=data, layout=layout )
iplot(fig)

```

```python
import plotly.plotly as py
import plotly.graph_objs as go

import pandas as pd

df = pd.read_csv('https://raw.githubusercontent.com/plotly/datasets/master/2011_us_ag_exports.csv')

for col in df.columns:
    df[col] = df[col].astype(str)

scl = [
    [0.0, 'rgb(242,240,247)'],
    [0.2, 'rgb(218,218,235)'],
    [0.4, 'rgb(188,189,220)'],
    [0.6, 'rgb(158,154,200)'],
    [0.8, 'rgb(117,107,177)'],
    [1.0, 'rgb(84,39,143)']
]

df['text'] = df['state'] + '<br>' + \
    'Beef ' + df['beef'] + ' Dairy ' + df['dairy'] + '<br>' + \
    'Fruits ' + df['total fruits'] + ' Veggies ' + df['total veggies'] + '<br>' + \
    'Wheat ' + df['wheat'] + ' Corn ' + df['corn']

data = [go.Choropleth(
    colorscale = scl,
    autocolorscale = False,
    locations = df['code'],
    z = df['total exports'].astype(float),
    locationmode = 'USA-states',
    text = df['text'],
    marker = go.choropleth.Marker(
        line = go.choropleth.marker.Line(
            color = 'rgb(255,255,255)',
            width = 2
        )),
    colorbar = go.choropleth.ColorBar(
        title = "Millions USD")
)]

layout = go.Layout(
    title = go.layout.Title(
        text = '2011 US Agriculture Exports by State<br>(Hover for breakdown)'
    ),
    geo = go.layout.Geo(
        scope = 'world',
        projection = go.layout.geo.Projection(type = 'albers usa'),
        showlakes = True,
        lakecolor = 'rgb(255, 255, 255)'),
)

fig = go.Figure(data = data, layout = layout)
py.iplot(fig, filename = 'd3-cloropleth-map')
```
