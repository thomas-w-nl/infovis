```python
from plotly.offline import init_notebook_mode, iplot
import plotly.graph_objs as go

# Start notebook mode
init_notebook_mode(connected=True)

import pandas as pd 
import numpy as np
```

```python
df = pd.read_csv('globalterrorismdb_0718dist.csv', encoding = "ISO-8859-1")


print(" ".join(df.keys()))

df.head(n=10) # eerste 10 van de in totaal 17 rijen
```

```python
data = go.Histogram(
    x=df['iyear']
)

layout = go.Layout(
    title='Number of attacks per year'
)

fig = go.Figure(data=[data], layout=layout)
iplot(fig)
```

```python
years = list(set(df['iyear']))

df_budget = pd.read_csv('military-expenditure-as-share-of-gdp.csv',  encoding = "ISO-8859-1")
us_budget = df_budget[df_budget['Code'] == 'USA']
us_budget = us_budget[us_budget['Year'] > 1970]


world_budget = [np.average(df_budget[df_budget['Year'] == year]['Expenditure']) for year in years]
df_budget.head(n=10)


# Create a trace
budget_trace = go.Scatter(
    x = us_budget['Year'],
    y = world_budget,
    yaxis='y2'
)


deaths_per_year = [np.nansum(df[df['iyear'] == year]['nkill']) for year in years]


data = go.Bar(
    x=years,
    y=deaths_per_year
)

layout = go.Layout(
    title='Number of deaths by terrorist attacks per year',
    yaxis=dict(
        title='yaxis title'
    ),
    yaxis2=dict(
        title='World budget as percent of gpb',
        overlaying='y',
        side='right'
    )
)



fig = go.Figure(data=[data, budget_trace], layout=layout)
iplot(fig)
```

```python
df = pd.read_csv('https://raw.githubusercontent.com/plotly/datasets/master/2014_world_gdp_with_codes.csv')

data = [go.Choropleth(
    locations = df['CODE'],
    z = df['GDP (BILLIONS)'],
    text = df['COUNTRY'],
    colorscale = [
        [0, "rgb(5, 10, 172)"],
        [0.35, "rgb(40, 60, 190)"],
        [0.5, "rgb(70, 100, 245)"],
        [0.6, "rgb(90, 120, 245)"],
        [0.7, "rgb(106, 137, 247)"],
        [1, "rgb(220, 220, 220)"]
    ],
    autocolorscale = False,
    reversescale = True,
    marker = go.choropleth.Marker(
        line = go.choropleth.marker.Line(
            color = 'rgb(180,180,180)',
            width = 0.5
        )),
    colorbar = go.choropleth.ColorBar(
        tickprefix = '$',
        title = 'GDP<br>Billions US$'),
)]

layout = go.Layout(
    title = go.layout.Title(
        text = '2014 Global GDP'
    ),
    geo = go.layout.Geo(
        showframe = False,
        showcoastlines = False,
        projection = go.layout.geo.Projection(
            type = 'equirectangular'
        )
    ),
    annotations = [go.layout.Annotation(
        x = 0.55,
        y = 0.1,
        xref = 'paper',
        yref = 'paper',
        text = 'Source: <a href="https://www.cia.gov/library/publications/the-world-factbook/fields/2195.html">\
            CIA World Factbook</a>',
        showarrow = False
    )]
)

fig = go.Figure(data = data, layout = layout)
iplot(fig, filename="wap.png")
```
